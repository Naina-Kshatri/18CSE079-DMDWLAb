```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
```

```
url="https://raw.githubusercontent.com/Naina-Kshatri/18CSE079-DMDWLAb/main/student-mat.csv"
df=pd.read_csv(url)
df.head()
```

```
X=df[['traveltime']]
Y=df[['studytime']]
print(X)
print(Y)
```

```
covariance=np.cov(X,Y,bias=True)[0][1]
print(covariance)
```
